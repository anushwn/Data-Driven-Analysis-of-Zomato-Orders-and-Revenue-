```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataframe=pd.read_csv("D:\zomato project\Zomato data .csv")
print(dataframe.head())
```

```python
dataframe
```

```python
let's convert the data type of the "rate" column to float and remove the denominator.
```

```python
def handleRate(value):
    value=str(value).split('/')
    value=value[0];
    return float(value)

dataframe['rate']=dataframe['rate'].apply(handleRate)
print(dataframe.head())
```

```python
Summary of the data frame
```

```python
dataframe.info()
```

```python
Conlusion - There is no NULL value in dataframe.
```

```python
Type of Resturant
```

```python
sns.countplot(x=dataframe['listed_in(type)'])
plt.xlabel("Type of restaurant")
```

```python
Conclusion : The majority of the restaurant fall into the dining category
```

```python
Dining restaurants are preferred by a larger number of individuals.
```

```python
grouped_data=dataframe.groupby("listed_in(type)")["votes"].sum()
result=pd.DataFrame({"votes":grouped_data})
plt.plot(result,c="green",marker="o")
plt.xlabel("Type of restaurant",c="red",size=20)
plt.ylabel("Votes",c="red",size=20)
```

```python
The majority of restaurants received ratings
```

```python
plt.hist(dataframe['rate'],bins=2)
plt.title("Ratings Distribution")
plt.show()
```

```python
Conslusion : The majority of restaurants received ratings ranging from 3.5 to 4
```

```python
couple_data=dataframe['approx_cost(for two people)']
sns.countplot(x=couple_data)
```

```python
The majority of couples prefer restaurants with an approximate cost of 300 rupees.
```

```python
Whether online orders receive higher ratings than offline orders
```

```python
plt.figure(figsize=(4,4))
sns.boxplot(x='online_order',y='rate',data=dataframe)
```

```python
Conslusion: Offline orders received lower ratings in comparison to online orders, which obtained excellent ratings.
```

```python
pivot_table=dataframe.pivot_table(index='listed_in(type)',columns='online_order', values='votes', aggfunc='sum')
sns.heatmap(pivot_table,annot=True,cmap="YlGnBu",fmt='d')
plt.title("Heatmap")
plt.xlabel("Online Order")  
plt.ylabel("Listed In (Type)")
plt.show()
```

```python
Conclusion : Dining restaurants primarily accept offine orders, whereas cafes primarily receive online orders.This suggests that clients prefer to place orders in person at restaurants, but prefer online ordering at cafes.
```
